```python
r2 = []
var = []
last = lastVar =  0
for n in range(1,10):
    model = LinearRegression()
    polyModel = PolynomialFeatures(degree=n, include_bias=False)
    poly_features = polyModel.fit_transform(x)
    model.fit(poly_features,y)
    y_predicted = model.predict(poly_features)
        
    plt.scatter(x,y,c = "k")
    plt.plot(x,y_predicted, c="b")
    print(np.var(np.abs(y - y_predicted)), r2_score(y, y_predicted))
    
    r2.append(r2_score(y, y_predicted)-last)
    var.append(np.var(np.abs(y - y_predicted))-lastVar)
    last = r2_score(y, y_predicted)    
    lastVar = np.var(np.abs(y - y_predicted))
```
